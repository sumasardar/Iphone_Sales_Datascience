Top 10 highest-rated iphone on Flipkart in India ?

```python
data.head()
```

```python
highest_rated = data.sort_values(by = ["Star Rating"], ascending = False)

highest_rated = highest_rated.head(10)
```

```python
print(highest_rated["Product Name"])
```

```python
print(highest_rated)
```

How many rating do the highest-rated iphones on Flipkart have ?

```python
data.head()
```

```python
iphones = highest_rated["Product Name"].value_counts()
label = iphones.index
counts = highest_rated["Number Of Ratings"]
figure = px.bar(highest_rated, x = label, y = counts)
figure.show()
```

Which iphones has the highest number of reviews on flipkart ?

```python
data.head()
```

```python
iphones = highest_rated["Product Name"].value_counts()
label = iphones.index
counts = highest_rated["Number Of Reviews"]
figure = px.bar(highest_rated, x = label, y = counts, title = "number of reviews of Rated iphone")
figure.show()
```

What is the relationship between the sale price of iphones and the number of rating on flipkart ?

```python
figure = px.scatter(data_frame=data, x="Number Of Ratings", y="Sale Price", size="Discount Percentage", trendline="ols", title="relationship between sales price and number of rating")

figure.show()
```

What is the relationship between the discount percentage and the number of rating on flipkart ?

```python
figure = px.scatter(data_frame = data, x = "Number Of Ratings",
                    y= "Discount Percentage", size = "Sale Price",
                    trendline="ols",
                    title= "Relationship between Discount and Number of Rating of iphone")

figure.show()
```

Most expensive and least expensive ?

```python
data.head()
```

```python
most_expensive = data.loc[data["Sale Price"].idxmax()]
least_expensive = data.loc[data["Sale Price"].idxmin()]

print("Most Expensive Product:")
print(most_expensive)

print("\nLeast Expensive Product:")
print(least_expensive)
```
